```python
import numpy as np
import cv2
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats
import depth_tools as tool
%matplotlib inline

mpl.rcParams['axes.xmargin'] = 0
mpl.rcParams['axes.ymargin'] = 0
```

```python
ROOT = '../../data/'
BOARD = ROOT + 'board/'
RENDER = ROOT + 'render_wave1_300/'
OUT = '../../output/check_shading-shape/'
size = 120
```

```python
fig, ax = plt.subplots(2, 4, figsize=(12,6))

ax[0, 0].set_title('Board-A')
ax[0, 0].axis('off')
ax[1, 0].set_ylim(-3, 3)

ax[0, 1].set_title('Board-B')
ax[0, 1].axis('off')
ax[1, 1].set_ylim(-3, 3)

ax[0, 2].set_title('Board-C')
ax[0, 2].axis('off')
ax[1, 2].set_ylim(-3, 3)

ax[0, 3].set_title('1wave')
ax[0, 3].axis('off')
ax[1, 3].set_ylim(-3, 3)

###################################################
DIR = BOARD
idx = 0
x = 800
y = 600
shade = cv2.imread(DIR + 'shading/shading{:03d}.bmp'.format(idx), 0)
shade = shade[y:y+size, x:x+size]
depth = cv2.imread(DIR + 'rec_ajusted/depth{:03d}.bmp'.format(idx), -1)
depth = tool.unpack_bmp_bgra_to_float(depth[y:y+size, x:x+size])
gt = cv2.imread(DIR + 'gt/gt{:03d}.bmp'.format(idx), -1)
gt = tool.unpack_bmp_bgra_to_float(gt[y:y+size, x:x+size])

slice_shade = shade[:, size//2]
slice_diff = depth[:, size//2] - gt[:, size//2]

shade_3ch = np.dstack([shade, shade, shade])
shade_3ch[:, size//2] = 255
shade_3ch = cv2.rotate(shade_3ch, cv2.ROTATE_90_COUNTERCLOCKWISE)

ax[0, 0].imshow(shade_3ch)
ax[1, 0].plot(range(size), scipy.stats.zscore(slice_shade))
ax[1, 0].plot(range(size), scipy.stats.zscore(slice_diff))
###################################################
idx = 41
x = 800
y = 600
shade = cv2.imread(DIR + 'shading/shading{:03d}.bmp'.format(idx), 0)
shade = shade[y:y+size, x:x+size]
depth = cv2.imread(DIR + 'rec_ajusted/depth{:03d}.bmp'.format(idx), -1)
depth = tool.unpack_bmp_bgra_to_float(depth[y:y+size, x:x+size])
gt = cv2.imread(DIR + 'gt/gt{:03d}.bmp'.format(idx), -1)
gt = tool.unpack_bmp_bgra_to_float(gt[y:y+size, x:x+size])

slice_shade = shade[:, size//2]
slice_diff = depth[:, size//2] - gt[:, size//2]

shade_3ch = np.dstack([shade, shade, shade])
shade_3ch[:, size//2] = 255
shade_3ch = cv2.rotate(shade_3ch, cv2.ROTATE_90_COUNTERCLOCKWISE)

ax[0, 1].imshow(shade_3ch)
ax[1, 1].plot(range(size), scipy.stats.zscore(slice_shade))
ax[1, 1].plot(range(size), scipy.stats.zscore(slice_diff))
###################################################
idx = 49
x = 800
y = 600
shade = cv2.imread(DIR + 'shading/shading{:03d}.bmp'.format(idx), 0)
shade = shade[y:y+size, x:x+size]
depth = cv2.imread(DIR + 'rec_ajusted/depth{:03d}.bmp'.format(idx), -1)
depth = tool.unpack_bmp_bgra_to_float(depth[y:y+size, x:x+size])
gt = cv2.imread(DIR + 'gt/gt{:03d}.bmp'.format(idx), -1)
gt = tool.unpack_bmp_bgra_to_float(gt[y:y+size, x:x+size])

slice_shade = shade[:, size//2]
slice_diff = depth[:, size//2] - gt[:, size//2]

shade_3ch = np.dstack([shade, shade, shade])
shade_3ch[:, size//2] = 255
shade_3ch = cv2.rotate(shade_3ch, cv2.ROTATE_90_COUNTERCLOCKWISE)

ax[0, 2].imshow(shade_3ch)
ax[1, 2].plot(range(size), scipy.stats.zscore(slice_shade))
ax[1, 2].plot(range(size), scipy.stats.zscore(slice_diff))
###################################################
DIR = RENDER
idx = 50
x = 800
y = 800
shade = cv2.imread(DIR + 'shade/{:05d}.png'.format(idx), 0)
shade = shade[y:y+size, x:x+size]
depth = cv2.imread(DIR + 'rec/{:05d}.bmp'.format(idx), -1)
depth = tool.unpack_bmp_bgra_to_float(depth[y:y+size, x:x+size])
gt = cv2.imread(DIR + 'gt/{:05d}.bmp'.format(idx), -1)
gt = tool.unpack_bmp_bgra_to_float(gt[y:y+size, x:x+size])

slice_shade = shade[:, size//2]
slice_diff = depth[:, size//2] - gt[:, size//2]

shade_3ch = np.dstack([shade, shade, shade])
shade_3ch[:, size//2] = 255
shade_3ch = cv2.rotate(shade_3ch, cv2.ROTATE_90_COUNTERCLOCKWISE)

ax[0, 3].imshow(shade_3ch)
ax[1, 3].plot(range(size), scipy.stats.zscore(slice_shade))
ax[1, 3].plot(range(size), scipy.stats.zscore(slice_diff))
###################################################
plt.savefig(OUT+'slice_compare.pdf')
plt.savefig(OUT+'slice_compare.svg')
plt.show()
```
