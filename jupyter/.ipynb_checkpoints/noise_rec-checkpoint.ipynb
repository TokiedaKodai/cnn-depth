```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import depth_tools
%matplotlib inline
```

```python
gt_img = cv2.imread('noise/00000.bmp', -1)
gt = depth_tools.unpack_bmp_bgra_to_float(gt_img)
```

```python
cam_params = {
    'focal_length': 0.037009,
    'pix_x': 1.25e-05,
    'pix_y': 1.2381443057539635e-05,
    'center_x': 790.902,
    'center_y': 600.635
}
```

```python
xyz_gt = depth_tools.convert_depth_to_coords(gt, cam_params)
depth_tools.dump_ply('noise/wave1.ply', xyz_gt.reshape(-1, 3).tolist())
```

```python
gt.shape
```

```python
h, w = gt.shape
```

```python
gt[600:610, 600:610]
```

```python
mean = 0
sigma = 0.0002
gauss = np.random.normal(mean, sigma, (h, w))
gauss = gauss.reshape(h, w)
```

```python
gauss[600:610, 600:610]
```

```python
gt_noise = gt + gauss
xyz_gt_noise = depth_tools.convert_depth_to_coords(gt_noise, cam_params)
depth_tools.dump_ply('noise/wave1_noise-0002.ply', xyz_gt_noise.reshape(-1, 3).tolist())
```
