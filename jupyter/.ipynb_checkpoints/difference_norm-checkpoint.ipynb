```python
import depth_tools
import cv2
import numpy as np
```

```python
src_dir = '../../data/render'

src_frame_dir = src_dir + '/proj'
src_gt_dir = src_dir + '/gt'
src_shading_dir = src_dir + '/shade'
src_rec_dir = src_dir + '/rec'

data_idx = 1

src_bgra = src_frame_dir + '/{:05d}.png'.format(data_idx)
src_depth_gt = src_gt_dir + '/{:05d}.bmp'.format(data_idx)
src_shading = src_shading_dir + '/{:05d}.png'.format(data_idx)
src_depth_gap = src_rec_dir + '/{:05d}.bmp'.format(data_idx)
```

```python
depth_img_gap = cv2.imread(src_depth_gap, -1)
depth_gap = depth_tools.unpack_bmp_bgra_to_float(depth_img_gap)

depth_img_gt = cv2.imread(src_depth_gt, -1)
depth_gt = depth_tools.unpack_bmp_bgra_to_float(depth_img_gt)
```

```python
depth_thre = 0.02
difference_threshold = 0.1
```

```python
depth_gap[600:610, 700:710]
```

```python
depth_gt[600:610, 700:710]
```

```python
difference = depth_gt - depth_gap
difference[600:610, 700:710]
```

```python
is_gt_available = depth_gt > depth_thre
is_gt_available[600:610, 700:710]
```

```python
len(is_gt_available)
```

```python
np.sum(is_gt_available)
```

```python
is_depth_close = np.logical_and(np.abs(difference) < difference_threshold,
                                is_gt_available)
np.sum(is_depth_close)
```

```python
mask = is_depth_close.astype(np.float32)
mask[600:610, 700:710]
```

```python
np.sum(mask)
```

```python
mean_difference = np.sum(difference * mask) / np.sum(mask)
mean_difference
```

```python
var_difference = np.sum(np.square((difference - mean_difference)*mask)) / np.sum(mask)
var_difference
```

```python
std_difference = np.sqrt(var_difference)
std_difference
```

```python
difference = (difference - mean_difference) / std_difference
difference[600:610, 700:710]
```

```python
difference *= mask
difference[600:610, 700:710]
```

```python
difference *= std_difference
difference[600:610, 700:710]
```

```python
depth_difference = difference + 1
depth_difference *= mask
depth_difference[600:610, 700:710]
```

```python
cam_params = {
    'focal_length': 0.037009,
    'pix_x': 1.25e-05,
    'pix_y': 1.2381443057539635e-05,
    'center_x': 790.902,
    'center_y': 600.635
}
```

```python
xyz = depth_tools.convert_depth_to_coords(depth_difference, cam_params)
depth_tools.dump_ply('difference{:03d}.ply'.format(data_idx), xyz.reshape(-1, 3).tolist())
```
