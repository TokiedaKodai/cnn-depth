```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DIR = '../../data/board/'
OUT = 'check_shading/'
size = 120
```

```python
ba0 = cv2.imread(DIR + 'shading/shading000.bmp', 0)
```

```python
ba0.shape
```

```python
pba0 = ba0[120:120+size, 1000:1000+size]
cv2.imwrite(OUT+'shading_board-A-0.png', pba0)
```

```python
spba0=pba0[:, size//2]
```

```python
plt.plot(range(len(spba0)), spba0)
plt.savefig(OUT+'slice_board-A-0.svg')
plt.show()
```
